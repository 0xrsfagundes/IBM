```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('automobileEDA.csv')

df.sort_index(axis=1, inplace=True)

df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

# Corr - calculate the correlation between columns of a DataFrame, using as defatult Pearson correlation coefficient

The coefficient ranges from -1 to 1:

 1: indicates a perfectly positive correlation
 0: indicates no correlation
-1: indicates a perfectly negative correlation

```python
df_corr = df.corr(numeric_only=True)

df_corr
```


# Heatmap

annot=True - adds the correlation coefficients to the heatmap
cmap='coolwarm' - sets the color scheme for the heatmap

annot_kws: dictionary of keyword arguments
    fontsize: Sets the font size of the annotations (e.g. {"fontsize": 12}).
    fontweight: Sets the font weight of the annotations (e.g. {"fontweight": "bold"}).
    color: Sets the color of the annotations (e.g. {"color": "red"}).
    ha: Sets the horizontal alignment of the annotations (e.g. {"ha": "center"}).
    va: Sets the vertical alignment of the annotations (e.g. {"va": "center"}).
    rotation: Sets the rotation angle of the annotations (e.g. {"rotation": 90}).
    bbox: Sets the bounding box properties of the annotations (e.g. {"bbox": {"facecolor": "white", "alpha": 0.5, "edgecolor": "red"}}).

```python
sns.heatmap(df_corr, annot=True, annot_kws={"fontsize":5}, cmap='coolwarm')
plt.show()
```


# Positive Linear Relationship


```python
df[['engine-size','price']].corr() #0.87
```

```python
# Engine size as potential predictor variable of price
sns.regplot(x="engine-size", y="price", data=df)

#The plt.ylim(0,) line of code sets the lower limit of the y-axis of the plot to 0. 
#This means that the y-axis will start at 0 and not display any negative values. 
#The upper limit of the y-axis is not specified, so it will adjust automatically to fit the data being plotted.

plt.ylim(0,)
```
