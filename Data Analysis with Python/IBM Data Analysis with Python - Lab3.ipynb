# Pearson Coefficient

<p>The Pearson Correlation measures the linear dependence between two variables X and Y.</p>
<p>The resulting coefficient is a value between -1 and 1 inclusive, where:</p>
<ul>
    <li><b>1</b>: Perfect positive linear correlation.</li>
    <li><b>0</b>: No linear correlation, the two variables most likely do not affect each other.</li>
    <li><b>-1</b>: Perfect negative linear correlation.</li>
</ul>


# P-value

<p>The p-value is a measure of the statistical significance of the correlation coefficient</p>
<p>Correlation significance:</p>
<ul>
    <li><b> $<$ 0.001 </b>:  strong evidence</li>
    <li><b> $<$ 0.05 </b>:  moderate evidence</li>
    <li><b> $<$ 0.1 </b>:  weak evidence</li>
    <li><b> $>$ 0.1 </b>:  no evidence</li>
</ul>

```python
from scipy import stats

# wheel-base vs price

pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("Pearson Correlation Coefficient:", pearson_coef, "\nP-value:", p_value)
```

# ANOVA: Analysis of Variance

<p>statistical method to test the significant differences between the means of two or more groups</p>

<p><b>F-test score</b>: 
<ul>
<li>ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score</li>
<li>A larger score means there is a larger difference between the means</li>
</ul>
</p>

<p><b>P-value</b>:  P-value tells how statistically significant our calculated score value is.</p>

```python
grouped_test2 = df[['drive-wheels', 'price']].groupby(['drive-wheels'])

grouped_test2.head()
```

```python
grouped_test2.get_group('4wd')['price'].to_frame()
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results:\nF=", f_val, "\nP =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results:\nF=", f_val, "\nP =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results:\nF=", f_val, "\nP =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])  
 
print( "ANOVA results:\nF=", f_val, "\nP =", p_val)    
```
