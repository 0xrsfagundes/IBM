```python
import pandas as pd

df = pd.read_excel('xls/Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)

df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)

df.head()
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent', 'RegName':'Region'}, inplace=True)

df.head()
```

```python
df['Total'] = df.sum(numeric_only=True, axis=1)

df.head()
```

```python
df.describe()
```

```python
df['Country']
```

```python
df[['Country',1980,1981,1982,1983,1984,1985]]
```

```python
df.set_index('Country',inplace=True)

df.head()
```

```python
df.index.name = None
```

```python
df.head()
```

```python
df_long = df
df_long.drop(['Total'], axis=1, inplace=True)
df_long.index.name = 'Country'
df_long.reset_index(inplace=True)
df_long = pd.melt(df, id_vars=['Country','Continent', 'Region', 'DevName'], var_name='Year', value_name='Value')
```

```python
df_long.head()
```

```python
import matplotlib.pyplot as plt

df_long[df_long['Country']=='Haiti'].plot(x='Year',y='Value')
plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.text(2000,6000,'2010 Earthquake')
plt.show()
```

```python
# assume df_long contains the data
df_chile = df_long[df_long['Country']=='Chile']
df_argentina = df_long[df_long['Country']=='Argentina']

# create the plot
ax = plt.gca()
df_chile.plot(x='Year', y='Value', ax=ax, label='Chile')
df_argentina.plot(x='Year', y='Value', ax=ax, label='Argentina')

# add labels and title
plt.title('Immigration from Chile and Argentina')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

# add the legend
ax.legend()

# show the plot
plt.show()
```

```python
# assume df_long contains the data
df_chile = df_long[df_long['Country']=='Chile']
df_brazil = df_long[df_long['Country']=='Brazil']

# create the plot with custom width
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
df_chile.plot(x='Year', y='Value', ax=ax, label='Chile')
df_brazil.plot(x='Year', y='Value', ax=ax, label='Brazil')

# add labels and title
plt.title('Immigration from Chile and Brazil')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

# add the legend
ax.legend()

# show the plot
plt.show()
```
